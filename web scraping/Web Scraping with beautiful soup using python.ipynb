```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

def getTags(tag):
  url = input("Enter a website URL: ")
  urlobj = urlopen(url)
  content = urlobj.read()
  soup = BeautifulSoup(content, "html.parser")
  return soup.findAll(tag)

def titleandmetaTags():
  url = input("Enter a website URL: ")
  urlobj = urlopen(url)
  content = urlobj.read()
  soup = BeautifulSoup(content, "html.parser")

  title = soup.title.string
  print("Title:", title)

  for meta in soup.find_all('meta'):
    if meta.has_attr('name'):
      if meta['name'].lower() in ['description', 'keywords']:
        print(meta.get('content'))

def allContent(url):
  urlobj = urlopen(url)
  content = urlobj.read()
  soup = BeautifulSoup(content, "html.parser")

  print("Paragraph Tags:")
  for p in getTags('p'):
    print(p)

  print("Heading Tags:")
  for tag in ['h1', 'h2', 'h3', 'h4', 'h5', 'h6']:
    print(f"{tag.upper()} Tags:")
    for heading in soup.find_all(tag):
      print(heading.text.strip())

def alt_tag():
  url = input("Enter a website URL: ")
  urlobj = urlopen(url)
  content = urlobj.read()
  urlobj.close()
  soup = BeautifulSoup(content, "html.parser")

  print("Extracting alt tags...")
  for img in soup.find_all('img', alt=True):
    print(img)

if __name__ == '__main__':
  titleandmetaTags()
  allContent(input("Enter another website URL for paragraphs and headings: "))
  alt_tag()
```
